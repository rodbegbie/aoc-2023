```python
# SAMPLE INPUT

INPUT = """R 6 (#70c710)
D 5 (#0dc571)
L 2 (#5713f0)
D 2 (#d2c081)
R 2 (#59c680)
D 2 (#411b91)
L 5 (#8ceee2)
U 2 (#caa173)
L 1 (#1b58a2)
U 2 (#caa171)
R 2 (#7807d2)
U 3 (#a77fa3)
L 2 (#015232)
U 2 (#7a21e3)"""
```

```python
# Part 1

rows = INPUT.splitlines()

x = 0
y = 0
borders = []

for row in rows:
    direction, steps, color = row.split()
    color = color[2:-1]
    steps = int(steps)
    for i in range(steps):
        borders.append((x,y))
        if direction == 'R':
            x += 1
        elif direction == 'L':
            x -= 1
        elif direction == 'U':
            y -= 1
        elif direction == 'D':
            y += 1

# Shoelace formula
area = 0
for i in range(len(borders) - 1):
    x1, y1 = borders[i]
    x2, y2 = borders[i + 1]
    area += x1 * y2 - x2 * y1

# Pick's theorem
perimeter = len(borders)
interior_area = abs(area) // 2 - perimeter // 2 + 1
print('FILL', interior_area + perimeter)
```
