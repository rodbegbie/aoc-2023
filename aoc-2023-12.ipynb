```python
# Part 1

rows = INPUT.splitlines()

def check_record(record, springs):
    found = []
    pieces  = record.split('.')
    for piece in pieces:
        if len(piece) > 0:
            found.append(len(piece))
    # print ("CHECKING", record, found, springs, found == springs)
    return found == springs

def count_arrangements(record, springs):
    # print ("RECORD", record)
    if "?" not in record:
        return 1 if check_record(record, springs) else 0
    record_hash = record[:].replace("?", "#", 1)
    record_dot = record[:].replace("?", ".", 1)

    return count_arrangements(record_hash, springs) + count_arrangements(record_dot, springs)
    
total = 0
for row in rows:
    record, springs_txt = row.split(' ')
    springs = [int(x) for x in springs_txt.split(',')]
    print (record, springs, count_arrangements(record, springs))
    total += count_arrangements(record, springs)

print ("TOTAL", total)

```

```python
# Part 2

rows = INPUT.splitlines()

def check_record(record, springs, strict=False):
    found = []
    pieces  = record.split('.')
    for piece in pieces:
        if len(piece) > 0:
            found.append(len(piece))

    found = tuple(found)
    if strict:
        return found == springs

    sub_match = True
    if len(found) > len(springs):
        return False
    if not found:
        return True
    for i in range(len(found[:-1])):
        if found[i] != springs[i]:
            sub_match = False
            break
    if found[len(found) - 1] > springs[len(found) - 1]:
        sub_match = False

    # print ("CHECKING", record, found, springs, sub_match)
    return sub_match

def count_arrangements(record, springs):
    # print ("RECORD", record)
    if "?" not in record:
        return 1 if check_record(record, springs, strict=True) else 0
    question_mark = record.find("?")
    record_hash = record[:].replace("?", "#", 1)
    record_dot = record[:].replace("?", ".", 1)
    hash_sub = record_hash[:question_mark + 1]
    dot_sub = record_dot[:question_mark + 1]
    while True:
        question_mark += 1
        if question_mark >= len(record) or record[question_mark] == "?":
            break
        hash_sub += record[question_mark]
        dot_sub += record[question_mark]
    hash_good = check_record(hash_sub, springs)
    dot_good = check_record(dot_sub, springs)

    if hash_good and dot_good:
        return count_arrangements(record_hash, springs) + count_arrangements(record_dot, springs)
    elif hash_good:
        return count_arrangements(record_hash, springs)
    elif dot_good:
        return count_arrangements(record_dot, springs)
    else:
        return 0
    
total = 0
FACTOR = 1
for row in rows:
    record, springs_txt = row.split(' ')
    record = "?".join([record] * FACTOR)
    springs = [int(x) for x in springs_txt.split(',')]
    springs = tuple(springs * FACTOR)
    answer = count_arrangements(record, springs)
    print (">>>>", record, springs, answer)
    total += answer
    
print ("TOTAL", total)

```
