```python
# Part 2

rows = INPUT.splitlines()

directions = rows[0]
current_nodes = []
nodes = {}

for row in rows[2:]:
    node, children = row.split(" = ")
    children = children.replace("(", "").replace(")", "").split(", ")
    nodes[node] = children
    if node[2] == "A":
        current_nodes.append(node)

from math import lcm

distances = []

for current_node in current_nodes:
    steps = 0
    while current_node[2] != "Z":
        # print ("NODE", current_node)
        direction = directions[steps % len(directions)]
        steps += 1
        if direction == "L":
            current_node = nodes[current_node][0]
        else:
            current_node = nodes[current_node][1]
    
    print ("NODE", current_node, "STEPS", steps)
    distances.append(steps)


print ("DISTANCES", distances, "LCM", lcm(*distances))
```
