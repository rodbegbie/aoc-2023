```python
# Part 1

# print (INPUT)

rows = INPUT.splitlines()
grid = [list(row) for row in rows]
seen = set()
visited = set()
beams = [(0, 0, 1, 0)]
while beams:
    print ("BEAMS", beams)
    new_beams = []
    for x, y, dx, dy in beams:
        # print ("X", x, "Y", y)
        if x < 0 or x >= len(grid[0]) or y < 0 or y >= len(grid):
            continue
        if (x,y,dx,dy) in visited:
            continue
        seen.add((x, y))
        visited.add((x, y, dx, dy))
        # print (grid[y][x], x, y, dx, dy)
        if grid[y][x] == '.':
            new_beams.append((x + dx, y + dy, dx, dy))
        elif grid[y][x] == '|':
            if dx == 0:
                new_beams.append((x, y + dy, dx, dy))
            else:
                new_beams.append((x, y-1, 0, -1))
                new_beams.append((x, y+1, 0, 1))
        elif grid[y][x] == '-':
            if dy == 0:
                new_beams.append((x + dx, y, dx, dy))
            else:
                new_beams.append((x-1, y, -1, 0))
                new_beams.append((x+1, y, 1, 0))
        elif grid[y][x] == '\\':
            new_beams.append((x+dy, y+dx, dy, dx))
        elif grid[y][x] == '/':
            new_beams.append((x-dy, y-dx, -dy, -dx))
    beams = new_beams

print ("ENERGIZED", len(seen))

```

```python
# Part 2

rows = INPUT.splitlines()
grid = [list(row) for row in rows]

def get_energized(starting_point):
    beams = [starting_point]
    seen = set()
    visited = set()
    while beams:
        # print ("BEAMS", beams)
        new_beams = []
        for x, y, dx, dy in beams:
            # print ("X", x, "Y", y)
            if x < 0 or x >= len(grid[0]) or y < 0 or y >= len(grid):
                continue
            if (x,y,dx,dy) in visited:
                continue
            seen.add((x, y))
            visited.add((x, y, dx, dy))
            # print (grid[y][x], x, y, dx, dy)
            if grid[y][x] == '.':
                new_beams.append((x + dx, y + dy, dx, dy))
            elif grid[y][x] == '|':
                if dx == 0:
                    new_beams.append((x, y + dy, dx, dy))
                else:
                    new_beams.append((x, y-1, 0, -1))
                    new_beams.append((x, y+1, 0, 1))
            elif grid[y][x] == '-':
                if dy == 0:
                    new_beams.append((x + dx, y, dx, dy))
                else:
                    new_beams.append((x-1, y, -1, 0))
                    new_beams.append((x+1, y, 1, 0))
            elif grid[y][x] == '\\':
                new_beams.append((x+dy, y+dx, dy, dx))
            elif grid[y][x] == '/':
                new_beams.append((x-dy, y-dx, -dy, -dx))
        beams = new_beams

    return len(seen)

max_ener = 0

for y in range(len(grid)):
    max_ener = max(max_ener, get_energized((0, y, 1, 0)))
    max_ener = max(max_ener, get_energized((len(grid[0])-1, y, -1, 0)))

for x in range(len(grid[0])):
    max_ener = max(max_ener, get_energized((x, 0, 0, 1)))
    max_ener = max(max_ener, get_energized((x, len(grid)-1, 0, -1)))

print ("MAX", max_ener)
```
