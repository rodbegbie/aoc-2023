```python
# Part 1

rows = INPUT.splitlines()

grid = []

queue = []

for row in rows:
    grid.append(list(row))

start_x = grid[0].index(".")
start_y = 0
queue.append((start_x, start_y, set()))


max_to_exit = 0

directions = [(0, 1, "v"), (0, -1, "^"), (1, 0, ">"), (-1, 0, "<")]

while queue:
    # print (queue)
    x, y, visited = queue.pop(0)
    if y == len(grid)-1:
        print(len(visited))
        max_to_exit = max(max_to_exit, len(visited))
        continue

    if (x, y) in visited:
        continue

    visited.add((x, y))

    for dx, dy, allowed_char in directions:
        new_x = x + dx
        new_y = y + dy
        if new_x < 0 or new_x >= len(grid[0]) or new_y < 0 or new_y >= len(grid):
            continue
        next_char = grid[new_y][new_x]
        if next_char == "#":
            continue
        if next_char == "." or next_char == allowed_char:
            queue.append((new_x, new_y, visited.copy()))

print ("LONGEST_PATH", max_to_exit)



```
