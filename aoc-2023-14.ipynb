```python
# SAMPLE INPUT

INPUT = """O....#....
O.OO#....#
.....##...
OO.#O....O
.O.....O#.
O.#..O.#.#
..O..#O..O
.......O..
#....###..
#OO..#...."""
```

```python
# Part 1

rows = INPUT.splitlines()
grid = [list(row) for row in rows]

total = 0
for i in range(len(grid[0])):
    pos = len(grid) 
    column = "".join([row[i] for row in grid])
    for chunk in column.split("#"):
        for j in range(chunk.count("O")):
            total += pos
            pos -= 1
        for j in range(chunk.count(".")):
            pos -= 1
        pos -= 1

print ("TOTAL", total)
```

```python
# Part 2

CYCLES = 1000000000
CYCLES = 4000

rows = INPUT.splitlines()
grid = [list(row) for row in rows]

for cycle in range(CYCLES):
    # print ("\nAfter cycle", cycle)
    # print ("\n".join("".join(row) for row in grid))
    for rotation in range(4):
        new_grid = []
        for col in range(len(grid[0])):
            new_grid.append(['?'] * len(grid))
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                new_grid[j][len(grid) - i - 1] = grid[i][j]
        grid = []
        # print ("NEW GRID", new_grid)
        for row_list in new_grid:
            row = "".join(row_list)
            chunks = row.split("#")
            new_row = []
            for chunk in chunks:
                new_row.extend(["."] * chunk.count("."))
                new_row.extend(["O"] * chunk.count("O"))
                new_row.extend(["#"])
            new_row = new_row[:-1]
            grid.append(new_row)


    total = 0
    for i in range(len(grid)):
        pos = len(grid) - i
        total += pos * ("".join(grid[i])).count("O")

    print ("CYCLE", cycle, "TOTAL", total)
```
