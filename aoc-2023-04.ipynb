```python
# Part 1

SCORES = (0, 1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024)

rows = INPUT.splitlines()

total = 0
for row in rows:
    (left, numbers_str) = row.split('|')
    (_, winning_numbers_str) = left.split(":")
    numbers = set(numbers_str.split())
    winning_numbers = set(winning_numbers_str.split())
    score = SCORES[len(numbers & winning_numbers)]
    total += score

print ("TOTAL", total)
```

```python
# Part 2

from collections import defaultdict

num_cards = defaultdict(lambda: 1)

total = 0
current_game = 0

rows = INPUT.splitlines()
for row in rows:
    current_game += 1
    total += num_cards[current_game]

    print ("CURRENT GAME", current_game, "NUM_CARDS", num_cards[current_game], "TOTAL", total)

    (left, numbers_str) = row.split('|')
    (_, winning_numbers_str) = left.split(":")
    numbers = set(numbers_str.split())
    winning_numbers = set(winning_numbers_str.split())
    wins = len(numbers & winning_numbers)

    for i in range(wins):
        num_cards[current_game + i + 1] += num_cards[current_game]
    

print ("TOTAL", total)
```
