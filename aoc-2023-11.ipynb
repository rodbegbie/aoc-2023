```python
# SAMPLE INPUT

INPUT = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#....."""
```

```python
# Part 1

galaxies = []
double_rows = set()
double_cols = set()
single_rows = set()
single_cols = set()

rows = INPUT.splitlines()
width = len(rows[0])
height = len(rows)
for y in range(height):
    for x in range(width):
        if rows[y][x] == "#":
            galaxies.append((x, y))
            single_cols.add(x)
            single_rows.add(y)

double_rows = set(range(height)) - single_rows
double_cols = set(range(width)) - single_cols

print ("GALAXIES", galaxies)
print ("DOUBLE ROWS", double_rows)
print ("DOUBLE COLS", double_cols)

total_distance = 0
for i in range(len(galaxies)):
    for j in range(i+1, len(galaxies)):
        x1, y1 = galaxies[i]
        x2, y2 = galaxies[j]
        distance = abs(x1 - x2) + abs(y1 - y2)
        distance += len(set(range(min(x1, x2), max(x1, x2))) & double_cols)
        distance += len(set(range(min(y1, y2), max(y1, y2))) & double_rows)
        total_distance += distance
        # print ("DISTANCE", i, j, distance, total_distance)


print ("TOTAL", total_distance)
```

```python
# Part 2

total_distance = 0
GROWTH_FACTOR = 1000000
for i in range(len(galaxies)):
    for j in range(i+1, len(galaxies)):
        x1, y1 = galaxies[i]
        x2, y2 = galaxies[j]
        distance = abs(x1 - x2) + abs(y1 - y2)
        distance += (len(set(range(min(x1, x2), max(x1, x2))) & double_cols) * (GROWTH_FACTOR - 1))
        distance += (len(set(range(min(y1, y2), max(y1, y2))) & double_rows) * (GROWTH_FACTOR - 1))
        total_distance += distance
        # print ("DISTANCE", i, j, distance, total_distance)


print ("TOTAL", total_distance)
```
