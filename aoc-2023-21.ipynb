```python
# SAMPLE INPUT

INPUT = """...........
.....###.#.
.###.##..#.
..#.#...#..
....#.#....
.##..S####.
.##..#...#.
.......##..
.##.#.####.
.##..##.##.
..........."""

STEPS = 6
```

```python
# Part 1
from collections import defaultdict

rows = INPUT.splitlines()

grid = []

queue = []
seen = set()

for row in rows:
    grid.append(list(row))
    if "S" in row:
        start = (row.index("S"), len(grid)-1)
        queue.append(start)
        seen.add(start)

i = 0
while i <= STEPS:
    # print (i, queue)
    new_queue = set()
    for (x, y) in queue:
        seen.add((x, y))
        for (dx, dy) in ((1, 0), (-1, 0), (0, 1), (0, -1)):
            new_x = x + dx
            new_y = y + dy
            if 0 <= new_x < len(grid[0]) and 0 <= new_y < len(grid) and (new_x, new_y) not in seen and grid[new_y][new_x] != "#":
                new_queue.add((new_x, new_y))
    queue = new_queue
    i += 1

could_be = set()
for (x, y) in seen:
    if not (x + y) % 2:
        could_be.add((x, y))

print (len(could_be))
```
