```python
# Part 2

rows = INPUT.splitlines()

cards = ("A", "K", "Q", "T", "9", "8", "7", "6", "5", "4", "3", "2", "J", )

hand_to_bid = {}

five_of_a_kind = []
four_of_a_kind = []
full_house = []
three_of_a_kind = []
two_pairs = []
one_pair = []
high_card = []

for row in rows:
    (hand, bid) = row.split()
    hand_to_bid[hand] = int(bid)

    sorted_hand = sorted(hand)
    # print ("HAND", hand, "".join(sorted_hand))
    if sorted_hand[0] == sorted_hand[1] == sorted_hand[2] == sorted_hand[3] == sorted_hand[4]:
        # print ("\tFUVE")
        five_of_a_kind.append(hand)
    elif sorted_hand[0] == sorted_hand[1] == sorted_hand[2] == sorted_hand[3] or sorted_hand[1] == sorted_hand[2] == sorted_hand[3] == sorted_hand[4]:
        if "J" in sorted_hand:
            five_of_a_kind.append(hand)
            # print ("\tFUVE (joker four)")
        else:
            # print ("\tFOUR")
            four_of_a_kind.append(hand)
    elif (sorted_hand[0] == sorted_hand[1] == sorted_hand[2] and sorted_hand[3] == sorted_hand[4]) or \
        (sorted_hand[0] == sorted_hand[1] and sorted_hand[2] == sorted_hand[3] == sorted_hand[4]):
        if "J" in sorted_hand:
            # print ("\tFIVE (joker full)")
            five_of_a_kind.append(hand)
        else:   
            # print ("\tFULL")
            full_house.append(hand)
    elif sorted_hand[0] == sorted_hand[1] == sorted_hand[2] or sorted_hand[1] == sorted_hand[2] == sorted_hand[3] or sorted_hand[2] == sorted_hand[3] == sorted_hand[4]:
        if "J" in sorted_hand:
            # print ("\tFOUR (joker three)")
            four_of_a_kind.append(hand)
        else:
            # print ("\tTHREE")
            three_of_a_kind.append(hand)
    elif (sorted_hand[0] == sorted_hand[1] and sorted_hand[2] == sorted_hand[3]) or \
        (sorted_hand[0] == sorted_hand[1] and sorted_hand[3] == sorted_hand[4]) or \
        (sorted_hand[1] == sorted_hand[2] and sorted_hand[3] == sorted_hand[4]):
        if "JJ" in "".join(sorted_hand):
            # print ("\tFOUR (joker 2p)")
            four_of_a_kind.append(hand)
        elif "J" in sorted_hand:
            # print ("\tFULL (joker 2p)")
            full_house.append(hand)
        else:
            # print ("\tTWOPAIR")
            two_pairs.append(hand)
    elif sorted_hand[0] == sorted_hand[1] or sorted_hand[1] == sorted_hand[2] or sorted_hand[2] == sorted_hand[3] or sorted_hand[3] == sorted_hand[4]:
        if "J" in sorted_hand:
            # print ("\tTHREE (joker 1p)")
            three_of_a_kind.append(hand)
        else:
            # print ("\tPAIR")
            one_pair.append(hand)
    else:
        if "J" in sorted_hand:
            # print ("\tPAIR (Joker high)")
            one_pair.append(hand)
        else:
            # print ("\tHIGH")
            high_card.append(hand)

total = 0
rank = 1

for hand_type in (high_card, one_pair, two_pairs, three_of_a_kind, full_house, four_of_a_kind, five_of_a_kind):
    hand_type.sort(key=lambda x: (cards.index(x[0]),cards.index(x[1]),cards.index(x[2]),cards.index(x[3]),cards.index(x[4])))
    hand_type.reverse()

    for hand in hand_type:
        total += rank * hand_to_bid[hand]
        rank += 1

print ("TOTAL", total)
```
