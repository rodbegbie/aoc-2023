```python
# Part 1

rows = INPUT.splitlines()

seeds = []
seed_to_soil_map = []
soil_to_fertilizer_map = []
fertilizer_to_water_map = []
water_to_light_map = []
light_to_temperature_map = []
temperature_to_humidity_map = []
humidity_to_location_map = []

current_map = None
for row in rows:
    if not row:
        current_map = None
        continue
    if row.startswith("seeds:"):
        seeds = [int(x) for x in row.split(":")[1].split()]
        continue
    if row == "seed-to-soil map:":
        current_map = seed_to_soil_map
        continue
    if row == "soil-to-fertilizer map:":
        current_map = soil_to_fertilizer_map
        continue
    if row == "fertilizer-to-water map:":
        current_map = fertilizer_to_water_map
        continue
    if row == "water-to-light map:":
        current_map = water_to_light_map
        continue
    if row == "light-to-temperature map:":
        current_map = light_to_temperature_map
        continue
    if row == "temperature-to-humidity map:":
        current_map = temperature_to_humidity_map
        continue
    if row == "humidity-to-location map:":
        current_map = humidity_to_location_map
        continue

    current_map.append([int(x) for x in row.split()])

locations = []
for seed in seeds:
    output = seed
    for map in (seed_to_soil_map, soil_to_fertilizer_map, fertilizer_to_water_map, water_to_light_map, light_to_temperature_map, temperature_to_humidity_map, humidity_to_location_map):
        for dest_range_start, source_range_start, range_length in map:
            if output in range(source_range_start, source_range_start + range_length):
                output += dest_range_start - source_range_start
                break
    locations.append(output)

print ("LOCATIONS", locations, "MIN", min(locations))
```

```python
# Part 2

rows = INPUT.splitlines()

seeds = []
seed_to_soil_map = []
soil_to_fertilizer_map = []
fertilizer_to_water_map = []
water_to_light_map = []
light_to_temperature_map = []
temperature_to_humidity_map = []
humidity_to_location_map = []

current_map = None
for row in rows:
    if not row:
        current_map = None
        continue
    if row.startswith("seeds:"):
        seed_ranges = [int(x) for x in row.split(":")[1].split()]
        while seed_ranges:
            seeds.append((seed_ranges[0], seed_ranges[1]))
            seed_ranges = seed_ranges[2:]
        continue
    if row == "seed-to-soil map:":
        current_map = seed_to_soil_map
        continue
    if row == "soil-to-fertilizer map:":
        current_map = soil_to_fertilizer_map
        continue
    if row == "fertilizer-to-water map:":
        current_map = fertilizer_to_water_map
        continue
    if row == "water-to-light map:":
        current_map = water_to_light_map
        continue
    if row == "light-to-temperature map:":
        current_map = light_to_temperature_map
        continue
    if row == "temperature-to-humidity map:":
        current_map = temperature_to_humidity_map
        continue
    if row == "humidity-to-location map:":
        current_map = humidity_to_location_map
        continue

    current_map.append([int(x) for x in row.split()])


map_source = seeds
for map in (seed_to_soil_map, soil_to_fertilizer_map, fertilizer_to_water_map, water_to_light_map, light_to_temperature_map, temperature_to_humidity_map, humidity_to_location_map):
    # print ("map_source", map_source)
    next_map_source = []
    while map_source:
        source_start, source_length = map_source.pop(0)
        mapped = False
        for dest_range_start, source_range_start, range_length in map:
            if mapped:
                break
            if source_start in range(source_range_start, source_range_start + range_length):
                if source_start + source_length in range(source_range_start, source_range_start + range_length):
                    # ALl move.
                    next_map_source.append((source_start + dest_range_start - source_range_start, source_length))
                    mapped = True
                    break
                else:
                    # Partial move.
                    num_moving = source_range_start + range_length - source_start
                    next_map_source.append((source_start + dest_range_start - source_range_start, num_moving))
                    map_source.append((source_start + num_moving, source_length - num_moving))
                    mapped = True
                    break
        if not mapped:
            next_map_source.append((source_start, source_length))
                
    map_source = next_map_source

print ("FINAL MAP SOURCE", map_source)

print ("FOUND!", sorted(map_source)[0][0])
```
