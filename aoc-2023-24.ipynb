```python
from __future__ import division 

# Part 1

rows = INPUT.splitlines()
lines = []


for row in rows:
    point, dir = row.split(" @ ")
    px, py, pz = map(int, point.split(", "))
    dx, dy, dz = map(int, dir.split(", "))
    lines.append((px, py, pz, dx, dy, dz))


def line(p1, p2):
    A = (p1[1] - p2[1])
    B = (p2[0] - p1[0])
    C = (p1[0]*p2[1] - p2[0]*p1[1])
    return A, B, -C

def intersection(L1, L2):
    D  = L1[0] * L2[1] - L1[1] * L2[0]
    Dx = L1[2] * L2[1] - L1[1] * L2[2]
    Dy = L1[0] * L2[2] - L1[2] * L2[0]
    if D != 0:
        x = Dx / D
        y = Dy / D
        return x,y
    else:
        return False

crossed = 0

for i in range(len(lines)):
    for j in range(i + 1, len(lines)):
        pxi, pyi, pzi, dxi, dyi, dzi = lines[i]
        pxj, pyj, pzj, dxj, dyj, dzj = lines[j]

        lip1 = (pxi, pyi)
        lip2 = (pxi + dxi, pyi + dyi)
        ljp1 = (pxj, pyj)
        ljp2 = (pxj + dxj, pyj + dyj)

        li = line(lip1, lip2)
        lj = line(ljp1, ljp2)

        intersect = intersection(li, lj)
        print (i, j, intersect)

        if not intersect:
            print ("> DON'T INTERSECT")
            continue

        if intersect[0] < MIN_XY or intersect[0] > MAX_XY or \
         intersect[1] < MIN_XY or intersect[1] > MAX_XY:
            print (">> INTERSECT OUTSIDE ZONE")
            continue

        if (intersect[0] < pxi and dxi > 0) or \
         (intersect[0] > pxi and dxi < 0) or \
         (intersect[1] < pyi and dyi > 0) or \
         (intersect[1] > pyi and dyi < 0) or \
         (intersect[0] < pxj and dxj > 0) or \
         (intersect[0] > pxj and dxj < 0) or \
         (intersect[1] < pyj and dyj > 0) or \
         (intersect[1] > pyj and dyj < 0) :
            print (">> INTESECT IN PAST")
            continue
        

        crossed += 1
        print (">>> CROSS IN FUTURE")


print ("CROSSED", crossed)
```
