```python
# Part 2

rows = INPUT.splitlines()

def get_grid_mirror(h_grid):
    v_grid = [""] * len(h_grid[0])
    for row in h_grid:
        for i in range(len(row)):
            v_grid[i] += row[i]
    # print ("\n".join(h_grid))
    # print ()
    # print ("\n".join(v_grid))

    for (multiple, grid) in ((100, h_grid), (1, v_grid)):
        all_good = 0
        smudge_fixed = False
        for i in range(len(grid)-1):
            smudge_fixed = False
            for j in range(len(grid)):
                if (i-j) < 0 or (i+j+1) >= len(grid):
                    if smudge_fixed:
                        all_good = i + 1    
                    break
                if grid[i-j] != grid[i+j+1]:
                    if not smudge_fixed:
                        diffs = 0
                        for k in range(len(grid[i-j])):
                            if grid[i-j][k] != grid[i+j+1][k]:
                                diffs += 1
                        if diffs == 1:
                            smudge_fixed = True
                            continue
                        else:
                            break
                    else:
                        break
            if all_good:
                break
        if all_good and smudge_fixed:
            print ("GOOD", all_good * multiple)
            return all_good * multiple

total = 0
grid = []
for row in rows:
    if row:
        grid.append(row)
        continue

    # break
    total += get_grid_mirror(grid)
    grid = []

total += get_grid_mirror(grid)
print ("TOTAL", total)

```
