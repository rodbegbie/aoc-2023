```python
# SAMPLE INPUT

INPUT = """467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598.."""
```

```python
# Part 1

from collections import defaultdict

rows = INPUT.splitlines()
numbers_seen = ["XXX"]
part_number_indexes = set()
index_locations = defaultdict(lambda: defaultdict(int))
current_number = ""

for row in range(len(rows)):
    for col in range(len(rows[0])):
        char = rows[row][col]
        if char.isdigit():
            current_number += char
            index_locations[row][col] = len(numbers_seen)
        else:
            if current_number:
                numbers_seen.append(int(current_number))
                current_number = ""

print (numbers_seen)
print (index_locations)

for row in range(len(rows)):
    for col in range(len(rows[0])):
        char = rows[row][col]
        if char != "." and not char.isdigit():
            # print(char, "part_number_indexes", part_number_indexes)
            for i in (-1, 0, 1):
                for j in (-1, 0, 1):
                    check_i = row + i
                    check_j = col + j

                    # print(check_i, check_j, index_locations[check_i][check_j])

                    if index_locations[check_i][check_j]:
                        part_number_indexes.add(index_locations[check_i][check_j])

total = 0
for x in part_number_indexes:
    total += numbers_seen[x]

print ("TOTAL", total)
```

```python
# Part 2

total = 0

for row in range(len(rows)):
    for col in range(len(rows[0])):
        char = rows[row][col]
        if char == "*":
            neighbours = set()
            # print(char, "part_number_indexes", total)
            for i in (-1, 0, 1):
                for j in (-1, 0, 1):
                    check_i = row + i
                    check_j = col + j

                    # print(check_i, check_j, index_locations[check_i][check_j])

                    if index_locations[check_i][check_j]:
                        neighbours.add(index_locations[check_i][check_j])
            if len(neighbours) == 2:
                neighbours = list(neighbours)
                total += numbers_seen[neighbours[0]] * numbers_seen[neighbours[1]]

print ("TOTAL", total)
```
